```python
# Importing libraries
import numpy as np
import pandas as pd
```

```python
# corn yield data

yield_data = pd.read_csv('https://github.com/edeliciouso/Predicting-Yields-for-Potatoes/blob/main/faostat_corn.csv?raw=true')
yield_data
```

```python
unique_values = yield_data['Area'].nunique()
print(f"Unique values in {'Area'}: {unique_values}")
```

```python
yield_data = yield_data.drop(['Domain Code', 'Domain', 'Area Code (M49)', 'Element Code', 'Element', 'Item Code (CPC)', 'Year Code', 'Unit', 'Flag', 'Flag Description', 'Note'], axis = 1)
yield_data
```

```python
# pesticides use data

pest_data  = pd.read_csv('https://github.com/edeliciouso/Predicting-Yields-for-Potatoes/blob/main/FAOSTAT_pest.csv?raw=true')
pest_data
```

```python
unique_values = pest_data['Area'].nunique()
print(f"Unique values in {'Area'}: {unique_values}")
```

```python
pest_data = pest_data.drop(['Domain Code', 'Domain', 'Area Code (M49)', 'Element Code', 'Element', 'Item Code', 'Item', 'Year Code', 'Unit', 'Flag', 'Flag Description'], axis = 1)
pest_data
```

```python
# Rainfall data

rain_data = pd.read_csv('https://github.com/edeliciouso/Predicting-Yields-for-Potatoes/blob/main/average-precipitation-per-year.csv?raw=true')
rain_data
```

```python
unique_values = rain_data['Entity'].nunique()
print(f"Unique values in {'Entity'}: {unique_values}")
```

```python
# modify the dataframe so it only shows the important ones

rain_data = rain_data.drop('Code', axis = 1)
rain_data = rain_data[rain_data['Year'] >= 2009]
rain_data.rename(columns={'Entity': 'Country'}, inplace=True) # rename entity to data
rain_data
```

```python
# fertilizer data

fert_data = pd.read_csv('https://github.com/edeliciouso/Predicting-Yields-for-Potatoes/blob/main/fertilizerData.csv?raw=true')
fert_data
```

```python
unique_values = fert_data['Entity'].nunique()
print(f"Unique values in {'Entity'}: {unique_values}")
```

```python
# fert_data = fert_data.drop(['Domain Code', 'Domain', 'Area Code (M49)', 'Element Code', 'Element', 'Item Code', 'Item', 'Year Code', 'Unit', 'Flag', 'Flag Description'], axis = 1)
# fert_data
```

```python
# rename the values

yield_data.rename({'Area': 'Country', 'Value': 'Yield (100 hg/ha)'}, axis = 1, inplace = True)
pest_data.rename({'Area': 'Country', 'Value': 'Pesticides (Tons)'}, axis = 1, inplace = True)
rain_data.rename({'Average precipitation in depth (mm per year)' : 'Rainfall (mm)'}, axis = 1, inplace = True)
fert_data.rename({'Entity': 'Country'}, axis = 1, inplace = True)
```

```python
# Merge the data

yield_pest = pd.merge(yield_data, pest_data, on=['Country', 'Year'], how='inner')
yield_pest_rain = pd.merge(yield_pest, rain_data, on=['Country', 'Year'], how = 'inner')
merged_data = pd.merge(yield_pest_rain, fert_data, on=['Country', 'Year'], how = 'inner')
merged_data
```

```python
# Convert 'Year' column to datetime type if it's not already
yield_data['Year'] = pd.to_datetime(yield_data['Year'], format='%Y')
fert_data['Year'] = pd.to_datetime(fert_data['Year'], format='%Y')
pest_data['Year'] = pd.to_datetime(pest_data['Year'], format='%Y')
rain_data['Year'] = pd.to_datetime(rain_data['Year'], format='%Y')

# Function to check if each country has entries for each year
def check_complete_data(df, dataset_name):
    result = df.groupby('Country')['Year'].apply(lambda x: all(year in x.dt.year.unique() for year in range(2009, 2020)))
    print(f"{dataset_name} Data:")
    print(result)
    print(f"Number of countries with incomplete data: {len(result) - result.sum()}\n")

# Check for each dataset
check_complete_data(yield_data, "Yield")
check_complete_data(fert_data, "Fertilizer")
check_complete_data(pest_data, "Pesticide")
check_complete_data(rain_data, "Rain")
```

```python
# Function to check if each country has entries for each year
def check_complete_data(df, dataset_name):
    result = df.groupby('Country')['Year'].apply(lambda x: all(year in x.dt.year.unique() for year in range(2009, 2020)))

    print(f"{dataset_name} Data:")
    incomplete_countries = result[result == False].index.tolist()

    if incomplete_countries:
        print("Countries with incomplete data:")
        print(incomplete_countries)
        print(f"Number of countries with incomplete data: {len(incomplete_countries)}")
    else:
        print("All countries have complete data.\n")

# Check for each dataset
check_complete_data(yield_data, "Yield")
check_complete_data(rain_data, "Rainfall")
check_complete_data(pest_data, "Pesticide")
```

```python
# List of countries to exclude
countries_to_exclude = ['Denmark', 'Sweden', 'Sudan', 'Sudan (former)', 'South Sudan', 'Netherlands Antilles (former)', 'Sint Maarten (Dutch part)']

# Filter out rows with specified countries
merged_data = merged_data[~merged_data['Country'].isin(countries_to_exclude)]

# Display the updated DataFrame
merged_data
```
