```python
# Importing libraries
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/fertilizer-per-hectare.csv')
df
```

```python
# Convert 'Year' column to datetime type
df['Year'] = pd.to_datetime(df['Year'], format='%Y')

# Filter data for the years 2009 to 2019 and update the original DataFrame
df = df[(df['Year'] >= '2009') & (df['Year'] <= '2019')]

# Display the updated DataFrame
df
```

```python
unique_values = df['Entity'].nunique()
print(f"Unique values in {'Entity'}: {unique_values}")
```

```python
df = df.drop(['Code'], axis = 1)
df
```

```python
# Convert 'Year' column to datetime type if it's not already
df['Year'] = pd.to_datetime(df['Year'], format='%Y')

# Function to check if each country has entries for each year
def check_complete_data(df, dataset_name):
    result = df.groupby('Entity')['Year'].apply(lambda x: all(year in x.dt.year.unique() for year in range(2009, 2020)))

    print(f"{dataset_name} Data:")
    incomplete_countries = result[result == False].index.tolist()

    if incomplete_countries:
        print("Countries with incomplete data:")
        print(incomplete_countries)
        print(f"Number of countries with incomplete data: {len(incomplete_countries)}")
    else:
        print("All countries have complete data.\n")

# Check for each dataset
check_complete_data(df, "Fertilizer")
```

```python
# List of countries to exclude
countries_to_exclude = ['Eritrea', 'Ethiopia', 'Qatar', 'Rwanda', 'Senegal', 'Yemen', 'Bahamas', 'Bermuda', 'Micronesia (FAO)', 'Nauru', 'Samoa', 'Sudan', 'Sudan (former)']

# Filter out rows with specified countries
df = df[~df['Entity'].isin(countries_to_exclude)]

# Display the updated DataFrame
df
```

```python
# Check for null values in the DataFrame
null_values = df.isnull().sum()

# Display the count of null values for each column
print(null_values)

# Show rows with null values
rows_with_null = df[df.isnull().any(axis=1)]
rows_with_null
```

```python
# Convert 'Year' column to datetime type by creating a copy of the DataFrame
df = df.copy()
df['Year'] = pd.to_datetime(df['Year'], format='%Y-%m-%d')

# Extract only the year from the 'Year' column
df['Year'] = df['Year'].dt.year

# Sum the values of nitrogen, phosphate, and potash for each row
df['Fertilizer (kg/ha)'] = df[['Nutrient nitrogen N (total) | 00003102 || Use per area of cropland | 005159 || Kilograms per hectare',
                               'Nutrient phosphate P2O5 (total) | 00003103 || Use per area of cropland | 005159 || Kilograms per hectare',
                               'Nutrient potash K2O (total) | 00003104 || Use per area of cropland | 005159 || Kilograms per hectare']].sum(axis=1)

# Drop the original columns
df = df.drop(['Nutrient nitrogen N (total) | 00003102 || Use per area of cropland | 005159 || Kilograms per hectare',
              'Nutrient phosphate P2O5 (total) | 00003103 || Use per area of cropland | 005159 || Kilograms per hectare',
              'Nutrient potash K2O (total) | 00003104 || Use per area of cropland | 005159 || Kilograms per hectare'], axis=1)

# Display the updated DataFrame
df
```

```python
# Save the merged DataFrame to a CSV file
df.to_csv('fertilizerData.csv', index=False)
```
